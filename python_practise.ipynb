```python
import numpy as np

def filter_by_bit(nparray, bit_index, bit_value="1"):
    """
    bit_index:
        0 = first bit
        -1 = last bit
        positive or negative indices allowed
    bit_value:
        "0" or "1"
    """
    # Extract the bit at the given index for all rows
    bits = np.array([row[1][bit_index] for row in nparray])

    # Build mask
    mask = (nparray[:, 0] == '1') & (bits == bit_value)

    return nparray[mask]

for bit_index in range(len(nparray[0, 1])):
    filtered = filter_by_bit(nparray, bit_index, "1")
    print("Bit index", bit_index, ":")
    for i in filtered:
        print(i)
```

```python
def calculate_parity(bits):
    bits_list = list(enumerate(bits))
    xor = [i for i,bit in bits_list if bit]
    parity = 0
    for i in xor:
        parity = parity ^ i
    return parity
calculate_parity(bits)
```
